```python
# --------------------------------------------------------------------------
# 🤖 display mensaje
# --------------------------------------------------------------------------
def display_mensaje(msj):
    print("\n" + "="*80)
    print("\033[1m --- " + msj + " --- \033[0m")
    print("="*80)
```

```python
# ------------------------------------------------------------------------------
# 🤖 diccionario builder
# ------------------------------------------------------------------------------

def dictionary_builder(columnas_categoricas, msj):

    _dict_unique_values_ = {}

    # recorremos las columnas categoricas para obtener sus valores
    display_mensaje(msj)
    # recorreos las columnas categoricas
    for col in columnas_categoricas.columns:
        # obtenemos los vals unicos
        unique_values = columnas_categoricas[col].unique()
        # asignamos los valores unicos al dict
        _dict_unique_values_[col] = columnas_categoricas[col].unique()


    # para mostrar el resultado del diccionario en pantalla
    # convertimos el diccionario a un conjunto de listas
    data = [[key] + list(value) for key, value in _dict_unique_values_.items()]

    # definimos los titulos de la tabla
    headers = ["Columna"] + [f"Valor único {i+1}" for i in range(len(max(_dict_unique_values_.values(), key=len)))]
    # imprimiendo
    print(tabulate(data, headers=headers, tablefmt="grid"))

    return _dict_unique_values_;
```

## 🔍 Verificación de integridad y consistencia

En este bloque se confirma que el escalamiento sea correcto revisando que la media sea cercana a 0 y la desviación estándar cercana a 1, y que no hayan quedado valores de nulos o valores infinitos.

```python
# ------------------------------------------------------------------------------
# Verificación de integridad y consistencia
# ------------------------------------------------------------------------------
def verificar_consistencia(df):
    display_mensaje("Verificación de integridad y consistencia")
    
    # comprobar nulos
    nulos = df.isnull().sum().sum()
    
    # comprobar valores infinitos
    infinitos = np.isinf(df.select_dtypes(include=np.number)).values.sum()
    
    # validar escalamiento solo en variables float64 que fueron estandarizadas
    cols_escaladas = ['edad', 'ingreso_mensual', 'antiguedad_cliente_meses', 'limite_credito', 
                      'saldo_promedio', 'porcentaje_utilizacion', 'num_transacciones_mes', 
                      'pagos_atrasados_12m', 'reclamos_12m']
    
    mean_check = df[cols_escaladas].mean().abs().mean()
    std_check = df[cols_escaladas].std().mean()

    print(f"Total de nulos: {nulos}")
    print(f"Total de infinitos: {infinitos}")
    print(f"Media promedio: {mean_check:.6f}")
    print(f"Desviación estándar promedio: {std_check:.6f}")
    
    if nulos == 0 and infinitos == 0 and abs(mean_check) < 0.1:
        print("\n✅ RESULTADO: Dataset consistente para modelación.")
    else:
        print("\n⚠️ ADVERTENCIA: Se detectaron inconsistencias en el dataset.")

verificar_consistencia(df_prep)
```

```python
# ------------------------------------------------------------------------------
# Algoritmo de ensamble
# ------------------------------------------------------------------------------

# definicion de x, y
_TARGET_ = 'default_12m'
X = df_prep.drop(columns=[_TARGET_, 'id_cliente'], errors='ignore') # caracteristicas
y = df_prep[_TARGET_]

# ------------------------------------------------------------------------------
# división de datos para validación 80 - entrenamiento / 20 - pruebas
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

df_final_preparado = df_prep.copy()


display_mensaje("🤖 Aplicando transformaciones...")
```

```python
# ------------------------------------------------------------------------------
# Preparación de K-Means y Configuración Visual
# ------------------------------------------------------------------------------

# configuración limpia
display_mensaje("🤖 Preparando K-Means")
print("🔨 Seleccionando solo las variables continuas y de comportamiento...")

'''
Seleccionando solo las variables continuas y de comportamiento.
Se excluyen las variables One-Hot porque no son ideales para K-Means
'''
cols_clustering = [
    'edad', 'ingreso_mensual', 'antiguedad_cliente_meses', 'limite_credito', 
    'saldo_promedio', 'porcentaje_utilizacion', 'num_transacciones_mes', 
    'pagos_atrasados_12m', 'reclamos_12m'
]
X_cluster = df_final_preparado[cols_clustering]

# mostrando las primeras filas de los datos escalados listos para agrupar
#display(pd.DataFrame(X_cluster, columns=cols_clustering).head())
```

## 🤖 Aplicación de K-Means final con k = 4

```python
# ------------------------------------------------------------------------------
# Aplicando K-Means (Modelo Final)
# ------------------------------------------------------------------------------
from sklearn.decomposition import PCA

# reduciendo a 2 componentes para poder graficar en un plano 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_cluster)

k_optimo = 4
display_mensaje("🤖 Entrenando K-Means con k=4")

kmeans_final = KMeans(n_clusters=k_optimo, random_state=42, n_init=10)
kmeans_final.fit(X_cluster)

# Asignamos la etiqueta del cluster a cada cliente en el dataframe original
df_final_preparado['Cluster_Conductual'] = kmeans_final.fit_predict(X_cluster)

# etiquetas finales al DataFrame
df_final_preparado["cluster_final"] = kmeans_final.labels_

'''
Nota sobre Centroides:
-> los datos de entrada ya están estandarizados desde la actividad 1
'''
centroides_pca = pca.transform(kmeans_final.cluster_centers_)
print("✔️ Modelo entrenado y etiquetas guardadas.")
```

## 🏛️ Perfil de clusters

```python
# ------------------------------------------------------------------------------
# Perfil de clusters
# ------------------------------------------------------------------------------
import matplotlib.ticker as mtick

display_mensaje("Perfil de los Clusters")
print("Valores > 0 indican que el cluster está por ENCIMA del promedio de la cartera.")
print("Valores < 0 indican que el cluster está por DEBAJO del promedio de la cartera.\n")

# asginación de nombres a cada cluster
#  sea mapean los numeros a nombres descriptivos para facilitar la lectura
nombres_estrategicos = {
    0: 'Leales Estabilizados',
    1: 'Inactivos Bajo Límite',
    2: 'Apalancados Crónicos',
    3: 'VIP Volátiles'
}

# nueva columna temporal para el reporte
df_final_preparado['Perfil_Estrategico'] = df_final_preparado['Cluster_Conductual'].map(nombres_estrategicos)

# Tabla de perfiles
cluster_profile = df_final_preparado.groupby('Perfil_Estrategico')[cols_clustering].mean()

# métricas clave de negocio
cluster_profile['Tasa_Default_Real (%)'] = df_final_preparado.groupby('Perfil_Estrategico')['default_12m'].mean() * 100
cluster_profile['Total_Clientes'] = df_final_preparado.groupby('Perfil_Estrategico').size()

# orden de columnas
cols_ordenadas = ['Total_Clientes', 'Tasa_Default_Real (%)'] + cols_clustering
cluster_profile = cluster_profile[cols_ordenadas]

# mostrando el resumen
'''
# se emplea el estilo background_gradient para crear un mapa de calor en la tabla
# rojo = alto, azul = bajo, blanco = cero
'''
print("Leyenda \n -------------")
print(" 🔴 = Alto \n 🔵 = Bajo \n ⚪ = Cero")
display(cluster_profile.round(2).style.background_gradient(cmap='coolwarm', axis=0))
```

## 🔍 Detección de combinaciones de alto riesgo

```python
# ------------------------------------------------------------------------------
# Detección de combinaciones de alto riesgo
# ------------------------------------------------------------------------------
import matplotlib.ticker as mtick

print("🔍 Analizando combinaciones de comportamiento...")

# Definiendo flags de comportamiento extremo 
df_final_preparado['Mora_Alta'] = df_final_preparado['pagos_atrasados_12m'] > 1.0
df_final_preparado['Utilizacion_Alta'] = df_final_preparado['porcentaje_utilizacion'] > 1.0

# Agrupando clientes por combinaciones extremas
patrones_riesgo = df_final_preparado.groupby(['Mora_Alta', 'Utilizacion_Alta'])['default_12m'].agg(
    Total_Casos='count',
    Tasa_Default='mean'
).reset_index()

# Porcentajes
patrones_riesgo['Tasa_Default'] = (patrones_riesgo['Tasa_Default'] * 100).round(1)

# Ordenando de menor a mayor para guiar el ojo de izquierda a derecha
patrones_riesgo = patrones_riesgo.sort_values('Tasa_Default', ascending=True).reset_index(drop=True)

# ------------------------------------------------------------------------------
# Grafica de combinaciones de comportamiento
# ------------------------------------------------------------------------------
# Promedio global para la línea de referencia
promedio_global = df_final_preparado['default_12m'].mean() * 100

plt.figure(figsize=(10, 6))

# paleta de colores dinamicos
colores_dinamicos = [COLOR_RIESGO if tasa > promedio_global else COLOR_SEGURO for tasa in patrones_riesgo['Tasa_Default']]

# Barplot
ax = sns.barplot(
    data=patrones_riesgo, 
    x=patrones_riesgo.index, 
    y='Tasa_Default', 
    palette=colores_dinamicos
)

# ---------------------------------------------------------
# Títulos
# ---------------------------------------------------------
plt.figtext(0.05, 1.02, "La mora es el detonante definitivo del impago", 
            fontsize=17, fontweight='bold', color=COLOR_TEXTO, ha='left')

plt.figtext(0.05, 0.99, "Independientemente de la utilización de la línea, presentar historial de mora dispara el riesgo por encima del 90%.", 
            fontsize=10, color=_color_gray, ha='left')

# ---------------------------------------------------------
# Formato de Ejes y Etiquetas
# ---------------------------------------------------------
# etiquetas del eje X limpias
etiquetas_x = []
for i, row in patrones_riesgo.iterrows():
    mora = "MORA ALTA" if row['Mora_Alta'] else "Mora Normal"
    util = "UTILIZACIÓN ALTA" if row['Utilizacion_Alta'] else "Utilización Normal"
    etiquetas_x.append(f"{util}\n+\n{mora}")

ax.set_xticks(range(len(etiquetas_x)))
ax.set_xticklabels(etiquetas_x, fontsize=10, color=COLOR_TEXTO, fontweight='500')

# Etiquetas en las barras
for p in ax.patches:
    ax.annotate(f"{p.get_height():.1f}%", 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', fontsize=12, fontweight='bold', color=COLOR_TEXTO,
                xytext=(0, 8), textcoords='offset points')

# ---------------------------------------------------------
# Línea de Promedio y Limpieza Visual
# ---------------------------------------------------------
# Línea de promedio global
ax.axhline(y=promedio_global, color=_Gray_, linestyle='--', linewidth=1.5, alpha=0.7)

# etiqueta del promedio
ax.text(x=2.65, y=promedio_global + 1.5, 
        s=f"Riesgo Promedio: {promedio_global:.1f}%", 
        fontsize=9, color=_Gray_, fontweight='bold', ha='left')

# Limpieza de ruido
sns.despine(left=True)
ax.get_yaxis().set_visible(False)
plt.ylabel("")
plt.xlabel("")

# Nota al pie
plt.figtext(0.05, -0.0, "Fuente: Análisis de interacciones del modelo MLOps | Riesgo Histórico", 
            fontsize=9, color=_color_gray, fontstyle='italic')

plt.tight_layout()
plt.savefig("figs/2-comportamiento-combinado.png", bbox_inches='tight', dpi=300)
plt.show()
```

Cuando un cliente presenta **historial de mora alta**, la probabilidad de default se eleva hasta **99.5%**, tanto en escenarios de utilización normal como de utilización alta, superando ampliamente el riesgo promedio de la cartera.

## 🔍 Análisis de capacidad predictiva

En este bloque se visualiza rápidamente cuáles son las variables de mayor influencia en el negocio de NovaCredit. Esto nos ayudará a entender mejor el dataset para el diseño del modelo.

```python
# ------------------------------------------------------------------------------
# 🔍 Análisis de Capacidad Predictiva
# ------------------------------------------------------------------------------
def analizar_predictoras(df):
    target = 'default_12m'
    
    print("📊 Generando Análisis de Fuerzas Predictivas...")
    
    # Coorelacion de pearson
    correlaciones = df.corr()[target].sort_values(ascending=False)
    correlaciones_plot = correlaciones.drop(target)
    
    # Limpieza de indices
    correlaciones_plot.index = correlaciones_plot.index.str.replace('_', ' ').str.title()
    
    # config
    plt.figure(figsize=(10, 7))
    COLOR_PROTECCION = '#2ECC71'
    COLOR_AUX = '#7F8C8D'
    
    # paleta de colores dinámicos
    colores_dinamicos = [COLOR_RIESGO if val > 0 else COLOR_PROTECCION for val in correlaciones_plot.values]
    
    # grafica de barras
    ax = sns.barplot(
        x=correlaciones_plot.values, 
        y=correlaciones_plot.index, 
        palette=colores_dinamicos
    )
    
    # títulos
    plt.figtext(0.05, 1.0, "Factores que impulsan y frenan la morosidad", 
                fontsize=17, fontweight='bold', color=COLOR_TEXTO, ha='left')
    plt.figtext(0.05, 0.93, "Las variables en rojo actúan como detonantes de riesgo. \n"
                            "Las variables en verde actúan como escudos protectores de la cartera.", 
                fontsize=11, color=COLOR_AUX, ha='left')

    # Etiquetas y eje central em cero
    ax.axvline(x=0, color=COLOR_TEXTO, linestyle='-', linewidth=1.5, alpha=0.3)
    
    # valores en barras
    for i, p in enumerate(ax.patches):
        valor = correlaciones_plot.values[i]
        
        # Si es positivo, el texto va a la derecha de la barra. Si es negativo, a la izquierda.
        if valor > 0:
            ax.annotate(f"+{valor:.2f}", 
                        (p.get_width(), p.get_y() + p.get_height() / 2.), 
                        ha='left', va='center', fontsize=10, fontweight='normal', color=COLOR_RIESGO,
                        xytext=(5, 0), textcoords='offset points')
        else:
            ax.annotate(f"{valor:.2f}", 
                        (p.get_width(), p.get_y() + p.get_height() / 2.), 
                        ha='right', va='center', fontsize=10, fontweight='normal', color=COLOR_PROTECCION,
                        xytext=(-5, 0), textcoords='offset points')

    # limpieza
    sns.despine(left=True, bottom=True)
    ax.get_xaxis().set_visible(False)
    ax.tick_params(axis='y', length=0, labelsize=11, colors=COLOR_TEXTO)
    plt.ylabel("")
    
    # Nota informativa
    plt.figtext(0.05, -0.05, "Fuente: Coeficientes de Correlación de Pearson respecto al objetivo default_12m", 
                fontsize=9, color=COLOR_AUX, fontstyle='italic')
    
    plt.tight_layout()
    plt.subplots_adjust(top=0.88)
    plt.subplots_adjust(right=1.2)
    plt.savefig('figs/5-factores-positivos-negativos.png', bbox_inches='tight', dpi=300)
    plt.show()
    
    # resultados
    print("\n🚨 TOP 3 DETONANTES DE RIESGO:")
    for var, val in correlaciones_plot.head(3).items():
        print(f"  - {var}: +{val:.2f}")
        
    print("\n🛡️ TOP 3 ESCUDOS PROTECTORES:")
    for var, val in correlaciones_plot.tail(3).sort_values().items():
        print(f"  - {var}: {val:.2f}")


    return correlaciones_plot

# analisis
correlaciones_plot = analizar_predictoras(df_prep)
```

### 📃 Interpretación

La gráfica permite identificar qué **variables explican el incumplimiento en la cartera de NovaCredit y en qué magnitud**, diferenciando entre **detonantes de riesgo** y **factores protectores**.

En primer lugar, se destaca que **los pagos atrasados en los últimos 12 meses** son, con mucha diferencia, **el principal impulsor de la morosidad**, con un peso muy superior al resto de variables. Esta relación confirma que **el default es esencialmente un fenómeno de comportamiento reiterado**, no un evento aislado ni aleatorio. En segundo término, los **reclamos registrados en 12 meses** también muestran una contribución positiva relevante al riesgo, sugiriendo que los conflictos operativos o insatisfacción del cliente suelen preceder al incumplimiento financiero.

Otras variables como **canal de adquisición web, saldo promedio, ingresos o tipo de producto préstamo** aportan al riesgo, pero con una influencia minima en comparación con la mora histórica. Esto indica que es necesario de priorizar indicadores conductuales por encima de datos demográficos.

Desde el lado protector, la gráfica muestra que **la antigüedad del cliente es el principal freno a la morosidad**. A mayor tiempo de relación con la empresa, menor probabilidad de default, lo que valida la importancia estratégica de **retener clientes históricos** y ofrecerles condiciones diferenciadas.

Esta gráfica respalda decisiones de negocio basadas en evidencia objetiva para **intervenir rápido ante señales de mora y fortalecer la relación con clientes de largo plazo**.

Este hallazgo es crítico debido a que estos perfiles **no necesariamente aparecen como morosos en una primera lectura tradicional**, pero representan una **fuerte amenaza**.
